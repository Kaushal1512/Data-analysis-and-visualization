```python
% matplotlib inline
% config InlineBackend.figure_format="svg"
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
from tqdm import tqdm
```

## Utility functions

```python
def pooledStd(A, B):
    s1, n1 = np.std(A), len(A)
    s2, n2 = np.std(B), len(B)
    spooled = (((n1-1)*(s1**2) + (n2-1)*(s2**2))/(n1+n2-2))**0.5
    return spooled


def tPooled(A, B):
    x1, s1, n1 = np.mean(A), np.std(A), len(A)
    x2, s2, n2 = np.mean(B), np.std(B), len(B)
    meanDiff = x1-x2
    spooled = pooledStd(A, B)
    tScore = meanDiff/(spooled*np.sqrt(1/n1 + 1/n2))
    return tScore

def effectSize(A,B):
    x1, s1, n1 = np.mean(A), np.std(A), len(A)
    x2, s2, n2 = np.mean(B), np.std(B), len(B)
    d = np.abs(x1-x2)/pooledStd(A, B)
    return d
```

## Normal population with same variance and same sample size

```python

runs=10**5
n1, n2 = 40, 40 

np.random.seed(5)

tScoresList = []

for i in tqdm(range(runs)):
    sample1 = np.random.normal(0,1,size=n1)
    sample2 = np.random.normal(0,1,size=n2)
    tScoresList.append(tPooled(sample1, sample2))

tScoresList=np.array(tScoresList)
xaxis = np.arange(np.min(tScoresList), np.max(tScoresList), 0.1)

plt.hist(tScoresList, bins=xaxis, density=True, label=r"Experimental")
plt.plot(xaxis, ss..pdf(xaxis, n1+n2-2), "k", label=r"Theoretical")
plt.legend()
plt.grid(True)
plt.xlabel('T')
plt.ylabel('Frequency')
plt.savefig("normal_1.png", format="png", dpi=300)
plt.show()
```
